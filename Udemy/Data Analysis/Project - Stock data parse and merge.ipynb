```python
def eps_ttm(eps):
    list_ttm = []
    list_eps = []
    last_eps = []
    eps_rec = 0
    num = 0
    flag = False
    for i in range(len(eps)):        
        if np.isnan(eps[i]) and flag:            
            list_eps.append(eps_rec)
        elif np.isnan(eps[i]) and not flag:                 
            list_eps.append(eps_rec)
        elif not np.isnan(eps[i]) and flag:
            last_eps[num] = round(eps[i],2)
            eps_rec =  round(eps[i],2)
            list_eps.append(eps_rec)            
            num = (num+1)%4
        else:
            last_eps.append(round(eps[i],2))
            eps_rec = round(eps[i],2)
            list_eps.append(eps_rec)
            if num == 3:
                flag = True
            num = (num+1)%4
        list_ttm.append(sum(last_eps)) 
    return [list_eps,list_ttm]

len(abt_eps)
```
