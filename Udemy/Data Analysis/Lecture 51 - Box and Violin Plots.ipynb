```python
#Regular
import numpy as np
import pandas as pd
from numpy.random import randn
from pandas import Series, DataFrame

#Stats
from scipy import stats

#Plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
url ='http://en.wikipedia.org/wiki/Box_plot'
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot([data1,data2])
```

```python
sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.boxplot(data1,vert=False)
```

```python
sns.boxplot(data1,vert=False,whis=np.inf)
```

```python
#normal Dist
data1 = stats.norm(0,5).rvs(100)

#Two gamma dist. Concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50) - 1,
                        -1*stats.gamma(5).rvs(50)])

#Box plot both data1 and data2
sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.violinplot([data1,data2])
```

```python
sns.violinplot(data2,bw=0.01)
```

```python
sns.violinplot(data1)
```

```python
sns.violinplot(data1,inner='stick')
```
